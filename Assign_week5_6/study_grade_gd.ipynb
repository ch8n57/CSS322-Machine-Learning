```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Datasets/dataset1.csv')
hours = np.array(dataset['hours'])
marks = np.array(dataset['marks'])
```

```python
print(hours, marks, sep='\n')
```

```python
m = 0
c = 0
learning_rate = 0.005
iterations = 100
n = len(hours)
X = hours
Y = marks
cost_history = []

for i in range(iterations):
    Y_pred = m*X + c
    cost = (1/n)*np.sum((Y - Y_pred)**2)
    cost_history.append(cost)
    
    dm = (-2/n)*np.sum(X*(Y - Y_pred))
    dc = (-2/n)*np.sum(Y - Y_pred)
    
    m = m - learning_rate*dm
    c = c - learning_rate*dc
```

```python
print("Gradient Descent m:", m)
print("Gradient Descent c:", c)
```

```python
plt.figure()
plt.plot(range(iterations), cost_history)
plt.xlabel("Iterations")
plt.ylabel("Cost (MSE)")
plt.title("Dataset 1 - Cost vs Iterations")
plt.show()
```

```python
plt.figure()
plt.scatter(X, Y)
plt.plot(X, m*X + c)
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.title("Dataset 1 - Final Regression Line")
plt.show()
```

```python
new_hours = 6
predicted_marks = m*new_hours + c
print("Prediction for 6 hours:", predicted_marks)
```
