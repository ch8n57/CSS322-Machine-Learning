```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Datasets/dataset2.csv')
size = np.array(dataset['house_size_1000_sqft'])
price = np.array(dataset['price_lakhs'])
```

```python
print(size, price, sep='\n')
```

```python
m = 0
c = 0
learning_rate = 0.005
iterations = 100
n = len(size)
X = size 
Y = price

cost_history = []

for i in range(iterations):
    Y_pred = m*X + c
    cost = (1/n)*np.sum((Y - Y_pred)**2)
    cost_history.append(cost)
    
    dm = (-2/n)*np.sum(X*(Y - Y_pred))
    dc = (-2/n)*np.sum(Y - Y_pred)
    
    m = m - learning_rate*dm
    c = c - learning_rate*dc
```

```python
print("Gradient Descent m:", m)
print("Gradient Descent c:", c)
```

```python
plt.figure()
plt.plot(range(iterations), cost_history)
plt.xlabel("Iterations")
plt.ylabel("Cost (MSE)")
plt.title("Dataset 2 - Cost vs Iterations")
plt.show()
```

```python
plt.figure()
plt.scatter(X, Y)
plt.plot(X, m*X + c)
plt.xlabel("House Size (1000 sq.ft)")
plt.ylabel("Price (Lakhs)")
plt.title("Dataset 2 - Final Regression Line")
plt.show()
```

```python
new_size = 2.7
predicted_price = m*new_size + c
print("Prediction for 2.7 (1000 sq.ft):", predicted_price)
```
