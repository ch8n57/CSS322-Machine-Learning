```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Datasets/dataset1.csv')
hours = np.array(dataset['hours'])
marks = np.array(dataset['marks'])
print(hours)
print(marks)
```

```python
x_mean = np.mean(hours)
y_mean = np.mean(marks)
print(x_mean)
print(y_mean)
```

```python
m = np.sum((hours - x_mean)*(marks - y_mean)) / np.sum((hours - x_mean)**2)
c = y_mean - m*x_mean
```

```python
print("Slope (m):", m)
print("Intercept (c):", c)
```

```python
plt.figure()
plt.scatter(hours, marks)
plt.plot(hours, m*hours + c)
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title("Dataset 1 - Linear Regression")
plt.show()
```

```python
new_hours = 6
predicted_marks = m*new_hours + c
print("Prediction for 6 hours:", predicted_marks)
```
