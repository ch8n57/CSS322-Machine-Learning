```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Datasets/dataset2.csv')
size = np.array(dataset['house_size_1000_sqft'])
price = np.array(dataset['price_lakhs'])
print(size, price, sep='\n')
```

```python
x_mean = np.mean(size)
y_mean = np.mean(price)
```

```python
print(x_mean, y_mean)
```

```python
m = np.sum((size - x_mean) * (price - y_mean)) / np.sum((size - x_mean) ** 2)
c = y_mean - m * x_mean
```

```python
print("Slope (m):", m)
print("Intercept (c):", c)
```

```python
plt.figure()
plt.scatter(size, price)
plt.plot(size, m*size + c)
plt.xlabel("House Size (1000 sq.ft)")
plt.ylabel("Price (Lakhs)")
plt.title("Dataset 2 - Linear Regression")
plt.show()
```

```python
new_size = 2.7
predicted_price = m*new_size + c
print("Prediction for 2.7 (1000 sq.ft):", predicted_price)
```
